### 14 ¿La cuenta total está relacionada con el tamaño? Justifica tu respuesta

Para ver si existe una relación se puede hacer revisando el promedio de la cuenta para cada tamaño de mesa.

```python
relacion=tips_data.groupby('size').agg({'total_bill':'mean'})
relacion
```

```python
import matplotlib.pyplot as plt
plt.plot(relacion)
plt.xlabel('Personas por mesa')
plt.ylabel('Promedio de la cuenta a pagar')
plt.grid()
```

## Respuesta

Como se puede ver en los datos de la tabla, el promedio aumenta a medida que el número de personas en la mesa también, al graficar los resultados se puede apreciar mucho más clara la tendencia a subir del promedio de las cuentas a medida que sube el número de personas, por lo tanto se concluye que **si hay una relación entre la cuenta total y el tamaño de la mesa.**
